```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("google/gemma-3-1b-it")
model = AutoModelForCausalLM.from_pretrained("google/gemma-3-1b-it")

# Set up a text-generation pipeline (CPU)
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
from transformers import pipeline

# Create a text generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Prompt for structured JSON output
sentence = "A woman with short hair, slightly fat, wearing a white top and black pants stood in front of the table, picked up a book from the table, and opened it to read."

prompt = f"""
Extract ONLY the NOUNS and VERBS from the following sentence: "{sentence}"
"""

# Generate text
outputs = generator(prompt, max_new_tokens=200, do_sample=False)
raw_output = outputs[0]['generated_text']


clean_output = raw_output.replace(prompt, "").strip()
print(clean_output)
```

```python
# Create a text generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

sentence = "A woman with short hair, slightly fat, wearing a white top and black pants stood in front of the table, picked up a book from the table, and opened it to read."

prompt = f"""
You are an information extraction system.
Analyze the following INPUT and output in strict JSON format only.

INPUT: "{sentence}"

The JSON schema must follow this structure:
[
  {{
    "id": "person_1",                # unique identifier for each entity
    "type": "person",                # type of entity (e.g., person, car, animal)
    "attributes": [ ... ],           # descriptive attributes (adjectives, appearance, clothing, etc.)
    "actions": [ ... ],              # verbs or actions performed
    "interactions": [ ... ]          # other entities they interact with
  }}
]

Now generate the JSON output:
"""

# Generate text
outputs = generator(prompt, max_new_tokens=300, do_sample=False)
raw_output = outputs[0]['generated_text']

# Clean it (remove prompt if repeated)
clean_output = raw_output.replace(prompt, "").strip()
print(clean_output)
```
